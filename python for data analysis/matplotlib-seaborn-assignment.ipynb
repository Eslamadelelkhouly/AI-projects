Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
figure = plt.figure()
ax1 = figure.add_axes([0,0,1,1])
ax2 = figure.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_ylabel('y')
ax1.plot(x,y)
ax2.plot(x,y)
figure
```

```python
fig = plt.figure()
ax = fig.subplots(ncols=2,nrows=1)
```

```python
for row in ax:
    row.plot(x,y)
    row.set_xlabel('x')
    row.set_ylabel('y')

ax[1].plot(x,y,color='red')
fig
```

# Seaborn Exercises

## Titanic Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

#### Recreate the plots below using the titanic dataframe.

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.displot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic_numeric = titanic.select_dtypes(include=[np.number])
sns.heatmap(titanic_numeric.corr(), cmap='coolwarm', annot=True, fmt=".2f")
```
